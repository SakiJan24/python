```python
from pyspark.sql import SparkSession

# Initialize a Spark session
spark = SparkSession.builder \
    .appName("CSV Reader") \
    .getOrCreate()


# Read the CSV file into a DataFrame
df = spark.read.csv('noDuplicatesIcfes.csv', header=True, inferSchema=True)

display(df)

```

### Revisando los tipos de datos de las columnas y también el tamaño del dataset

```python
df.dtypes
```

```python
# Obtener el número de filas en el DataFrame
num_rows = df.count()

# Obtener el número de columnas en el DataFrame
num_columns = len(df.columns)

# Imprimir el número de filas y columnas
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

### Conteo de nulos por columna

```python
# Count the total number of rows
total_rows = df.count()

# Count the number of distinct rows
distinct_rows = df.distinct().count()

# Check for duplicates
duplicates = total_rows - distinct_rows

print(f"Total rows: {total_rows}")
print(f"Distinct rows: {distinct_rows}")
print(f"Duplicate rows: {duplicates}")
df_no_duplicates = df.dropDuplicates()

# Verify the count of rows after removing duplicates
new_total_rows = df_no_duplicates.count()

print(f"Total rows after removing duplicates: {new_total_rows}")
```

```python
from pyspark.sql.functions import col, sum

# Print statement
print("Nulos por columna dentro del dataset de ICFES")

# Create a list to store the null counts
null_counts = []

# Iterate over each column to count null values
for col_name in df.columns:
    # Count the null values in the column
    null_count = df.select(col(col_name)).where(col(col_name).isNull()).count()
    # Append the null count to the list
    null_counts.append((col_name, null_count))

# Convert the null counts list to a PySpark DataFrame
null_counts_df = spark.createDataFrame(null_counts, ["Column Name", "Null Count"])

# Display the DataFrame
null_counts_df.show(n=51, truncate=False)
```

#### Eliminación del año 20201 ya que hubo muy pocas pruebas debido a que ese fin de semana se empezó la cuarentan en muchas ciudades del país

```python
from pyspark.sql.functions import col

# Valores a eliminar
values_to_remove = [20201]

# Filtrar el DataFrame para eliminar las filas donde '_c0' (PERIODO) está en values_to_remove
df_filtered = df.filter(~col('PERIODO').isin(values_to_remove))

# Obtener y mostrar los valores únicos de la columna '_c0' (PERIODO)
unique_periods = df_filtered.select('PERIODO').distinct().collect()

# Imprimir los valores únicos de '_c0'
print([row['PERIODO'] for row in unique_periods])
```

```python
# Definir las características categóricas y numéricas
from pyspark.ml.feature import OneHotEncoder
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator
print 
categorical_features = [ 'FAMI_ESTRATOVIVIENDA', 'COLE_AREA_UBICACION', 'COLE_BILINGUE', 'ESTU_GENERO']
numeric_features = ['EDAD']


# Convertir las características categóricas en índices utilizando StringIndexer y luego codificar en vectores con OneHotEncoder
indexers = [StringIndexer(inputCol=col, outputCol=col+"_index", handleInvalid="keep") for col in categorical_features]
encoder = OneHotEncoder(inputCols=[col+"_index" for col in categorical_features], outputCols=[col+"_encoded" for col in categorical_features])

# Armar todas las características en un solo vector
assembler = VectorAssembler(inputCols=numeric_features + [col+"_encoded" for col in categorical_features], outputCol="features")

# Definir el modelo de regresión logística
lr = LogisticRegression(featuresCol='features', labelCol='BINARY_PUNT_GLOBAL')

# Construir el pipeline
pipeline = Pipeline(stages=indexers + [encoder] + [assembler] + [lr])

# Dividir los datos en entrenamiento y prueba
df_selected = df_selected.na.fill(0)
train_df, test_df = df_selected.randomSplit([0.8, 0.2], seed=42)

# Entrenar el modelo
model = pipeline.fit(train_df)

# Hacer predicciones
predictions = model.transform(test_df)

# Evaluar el modelo
evaluator = BinaryClassificationEvaluator(labelCol='BINARY_PUNT_GLOBAL')
auc = evaluator.evaluate(predictions)

# Imprimir el área bajo la curva ROC
print("AUC:", auc)

# Matriz de confusión
y_true = predictions.select(['BINARY_PUNT_GLOBAL']).collect()
y_pred = predictions.select(['prediction']).collect()
conf_matrix = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
plt.title('Confusion Matrix')
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Curva ROC
fpr, tpr, _ = roc_curve(y_true, y_pred)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
